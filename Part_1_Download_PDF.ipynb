```python
df_url = pd.read_csv('urls.csv')
urls = df_url['URL']
```

```python
documents = []
file_contents=[]
URLs=[]

for i in urls:
    try:
        response = requests.get(i)
        file = open("data"+'\/'+i[-10:]+".pdf", "wb")
        file.write(response.content)
        file.close()

        pdf = pdfplumber.open("data"+'\/'+i[-10:]+".pdf")
        pages = pdf.pages
        pages_content=""
        for j in pages:
            text = j.extract_text()
            pages_content=pages_content+text
        pdf.close()
        file_contents.append(pages_content)
        URLs.append(i)

        pd.DataFrame({
                  'URL':URLs,
                  'file_contents': file_contents}).to_csv('contents' + '.csv', encoding='utf8', index = False)
    except: 
        print(i)
```
